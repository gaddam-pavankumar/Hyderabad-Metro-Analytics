```python
df.info()
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day_name()
df["Weekend"] = df["Day"].isin(["Saturday", "Sunday"])
```

```python
df.head()
```

```python
monthly = df.groupby("Month")["Passengers"].sum()

monthly
```

```python
import matplotlib.pyplot as plt

monthly.plot(kind="line", marker="o")

plt.title("Monthly Ridership Trend")
plt.xlabel("Month")
plt.ylabel("Passengers")
plt.show()
```

```python
daily_ridership = (
    df.groupby("Date")["Passengers"]
      .sum()
      .reset_index()
)

daily_ridership.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

plt.plot(
    daily_ridership["Date"],
    daily_ridership["Passengers"]
)

plt.title("Daily Metro Ridership")
plt.xlabel("Date")
plt.ylabel("Passengers")

plt.show()
```

```python
forecast_df = daily_ridership.rename(
    columns={
        "Date":"ds",
        "Passengers":"y"
    }
)

forecast_df.head()
```

```python
from prophet import Prophet

model = Prophet()

model.fit(forecast_df)
```

```python
future = model.make_future_dataframe(
    periods=30
)

forecast = model.predict(future)
```

```python
forecast[
    ["ds","yhat"]
].tail(30)
```

```python
model.plot(forecast)

plt.show()
```

```python
forecast.to_csv(
    "../data/forecast_results.csv",
    index=False
)
```
